```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, classification_report
import numpy as np

# Load the cleaned dataset
df = pd.read_csv('C:\\Users\\ronald\\Downloads\\blood_cancer_prediction_data_extended.csv')
# Features and target
X = df.drop(columns=['Genetic Mutations'])
y = df['Genetic Mutations']

# Split the data into training and test sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Initialize and train the Random Forest Classifier
rf_model = RandomForestClassifier(n_estimators=100, random_state=42)
rf_model.fit(X_train, y_train)

# Define a function to get user input with range validation
def get_user_input():
    try:
        # Age validation
        age = int(input("Enter Age: "))
        if not (0 <= age <= 120):
            raise ValueError("Age must be between 0 and 120.")

        # Gender validation
        gender = int(input("Enter Gender (0 for Male, 1 for Female): "))
        if gender not in [0, 1]:
            raise ValueError("Gender must be 0 (Male) or 1 (Female).")

        # Family History of Cancer validation
        family_history = int(input("Family History of Cancer (0 for No, 1 for Yes): "))
        if family_history not in [0, 1]:
            raise ValueError("Family history must be 0 (No) or 1 (Yes).")

        # WBC count validation
        wbc = float(input("Enter WBC Count (cells/µL): "))
        if not (4000 <= wbc <= 11000):
            raise ValueError("WBC count must be between 4,000 and 11,000 cells/µL.")

        # RBC count validation
        rbc = float(input("Enter RBC Count (million cells/µL): "))
        if not (4.2 <= rbc <= 6.1):
            raise ValueError("RBC count must be between 4.2 and 6.1 million cells/µL.")

        # Platelet count validation
        platelet = int(input("Enter Platelet Count (/µL): "))
        if not (150000 <= platelet <= 450000):
            raise ValueError("Platelet count must be between 150,000 and 450,000 /µL.")

        # Hemoglobin level validation
        hemoglobin = float(input("Enter Hemoglobin Level (g/dL): "))
        if not (12 <= hemoglobin <= 17.5):
            raise ValueError("Hemoglobin level must be between 12 and 17.5 g/dL.")

        # Lymphocyte count validation
        lymphocyte = float(input("Enter Lymphocyte Count (cells/µL): "))
        if not (1000 <= lymphocyte <= 4800):
            raise ValueError("Lymphocyte count must be between 1,000 and 4,800 cells/µL.")

        # Previous Cancer Diagnosis validation
        prev_cancer = int(input("Previous Cancer Diagnosis (0 for No, 1 for Yes): "))
        if prev_cancer not in [0, 1]:
            raise ValueError("Previous Cancer Diagnosis must be 0 (No) or 1 (Yes).")

        # Format the input into an array
        user_input = np.array([[age, gender, family_history, wbc, rbc, platelet, hemoglobin, lymphocyte, prev_cancer]])

        return user_input

    except ValueError as e:
        print(f"Invalid input: {e}")
        return None

# Get user input
user_data = get_user_input()
if user_data is not None:
    # Make a prediction
    prediction = rf_model.predict(user_data)

    # Output the prediction
    if prediction == 1:
        print("High Risk of Blood Cancer")
    else:
        print("Low Risk of Blood Cancer")

# Optional: Evaluate the model performance on the test set
y_pred = rf_model.predict(X_test)
print("\nModel Evaluation:")
print("Accuracy Score:", accuracy_score(y_test, y_pred))
print("\nClassification Report:")
print(classification_report(y_test, y_pred))
```

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, classification_report
import numpy as np

# Load dataset
df = pd.read_excel('C:\\Users\\ronald\\Downloads\\combined_file.xlsx')

# Prepare the feature set (X) and target (y)
X = df[['Age', 'Gender', 'Family History of Cancer', 'WBC Count', 'RBC Count', 
        'Platelet Count', 'Hemoglobin Level', 'Lymphocyte Count', 'Previous Cancer Diagnosis']]
y = df['Genetic Mutations'] # Assuming this represents blood cancer risk

# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Train a RandomForestClassifier model
rf_model = RandomForestClassifier(n_estimators=100, random_state=42)
rf_model.fit(X_train, y_train)

# Adjusted risk factors based on blood cancer and typical blood work ranges
def assess_risk(age, wbc, rbc, hemoglobin, platelet, lymphocyte):
    # Age risk adjustment
    if age <= 30:
        age_risk = "Low"
    elif 31 <= age <= 50:
        age_risk = "Moderate"
    else:
        age_risk = "High"
    
    # WBC count risk assessment based on blood cancer risk
    wbc_risk = "High" if wbc < 4500 or wbc > 11000 else "Low"
    
    # RBC count risk assessment
    rbc_risk = "High" if rbc < 4.5 or rbc > 6.0 else "Low"

    # Hemoglobin risk assessment
    hemoglobin_risk = "High" if hemoglobin < 12 or hemoglobin > 17 else "Low"

    # Platelet count risk assessment
    platelet_risk = "High" if platelet < 150000 or platelet > 400000 else "Low"

    # Lymphocyte count risk assessment
    lymphocyte_risk = "High" if lymphocyte < 1000 or lymphocyte > 4800 else "Low"
    
    return {
        "age_risk": age_risk,
        "wbc_risk": wbc_risk,
        "rbc_risk": rbc_risk,
        "hemoglobin_risk": hemoglobin_risk,
        "platelet_risk": platelet_risk,
        "lymphocyte_risk": lymphocyte_risk
    }

# Define a function to get user input and validate
def get_user_input_and_risk():
    try:
        age = int(input("Enter Age: "))
        if not (0 <= age <= 120):
            raise ValueError("Age must be between 0 and 120.")
        
        gender = int(input("Enter Gender (0 for Male, 1 for Female): "))
        if gender not in [0, 1]:
            raise ValueError("Gender must be 0 (Male) or 1 (Female).")
        
        family_history = int(input("Family History of Cancer (0 for No, 1 for Yes): "))
        if family_history not in [0, 1]:
            raise ValueError("Family History must be 0 (No) or 1 (Yes).")
        
        wbc = float(input("Enter WBC Count (cells/µL): "))
        if not (0 <= wbc <= 20000):
            raise ValueError("WBC count must be between 0 and 20,000 cells/µL.")
        
        rbc = float(input("Enter RBC Count (million cells/µL): "))
        if not (0 <= rbc <= 10):
            raise ValueError("RBC count must be between 0 and 10 million cells/µL.")
        
        platelet = int(input("Enter Platelet Count (/µL): "))
        if not (50000 <= platelet <= 1000000):
            raise ValueError("Platelet count must be between 50,000 and 1,000,000 /µL.")
        
        hemoglobin = float(input("Enter Hemoglobin Level (g/dL): "))
        if not (0 <= hemoglobin <= 25):
            raise ValueError("Hemoglobin level must be between 0 and 25 g/dL.")
        
        lymphocyte = float(input("Enter Lymphocyte Count (cells/µL): "))
        if not (0 <= lymphocyte <= 10000):
            raise ValueError("Lymphocyte count must be between 0 and 10,000 cells/µL.")
        
        prev_cancer = int(input("Previous Cancer Diagnosis (0 for No, 1 for Yes): "))
        if prev_cancer not in [0, 1]:
            raise ValueError("Previous Cancer Diagnosis must be 0 (No) or 1 (Yes).")
        
        # Format the input into an array for prediction
        user_input = np.array([[age, gender, family_history, wbc, rbc, platelet, hemoglobin, lymphocyte, prev_cancer]])

        # Assess risk levels based on ranges
        risk_factors = assess_risk(age, wbc, rbc, hemoglobin, platelet, lymphocyte)

        return user_input, risk_factors

    except ValueError as e:
        print(f"Invalid input: {e}")
        return None, None

# Get user input and risk assessment
user_data, risk_factors = get_user_input_and_risk()
if user_data is not None:
    # Make a prediction for blood cancer risk
    prediction_proba = rf_model.predict_proba(user_data)[0][1]  # Probability of blood cancer (class 1)

    # Output the prediction with percentage
    print(f"\nPredicted risk of blood cancer: {prediction_proba * 100:.2f}%")

    # Print risk factors based on ranges
    print("\nRisk Levels Based on Input:")
    for factor, risk in risk_factors.items():
        print(f"{factor}: {risk}")

# Optional: Evaluate the model performance on the test set
y_pred = rf_model.predict(X_test)
print("\nModel Evaluation:")
print("Accuracy Score:", accuracy_score(y_test, y_pred))
print("\nClassification Report:")
print(classification_report(y_test, y_pred))


```

```python
import pandas as pd
from fpdf import FPDF

# Function to generate PDF from CSV file
def csv_to_pdf(csv_file, pdf_file_name):
    # Load CSV data into DataFrame
    df = pd.read_csv(csv_file)

    # Initialize PDF
    pdf = FPDF()
    pdf.set_font("Arial", size=8)
    pdf.add_page()

    # Define column widths (Adjust as per your need)
    col_widths = [20, 20, 35, 20, 20, 20, 30, 25, 40, 30, 30, 30, 40, 40]
    
    # Add table header
    for i, column in enumerate(df.columns):
        pdf.cell(col_widths[i], 10, str(column), border=1)

    pdf.ln()  # Move to the next line

    # Add table rows
    for row in df.itertuples(index=False):
        for i, value in enumerate(row):
            pdf.cell(col_widths[i], 10, str(value), border=1)
        pdf.ln()

    # Save the generated PDF to the specified file
    pdf.output(pdf_file_name)

# File paths
csv_file_path = 'generated_data.csv'  # Your CSV file path
pdf_file_name = 'generated_data.pdf'  # Output PDF file name

# Convert CSV to PDF
csv_to_pdf(csv_file_path, pdf_file_name)

print(f"PDF file '{pdf_file_name}' generated successfully!")
```
